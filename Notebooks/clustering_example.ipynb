```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Clustering Techniques

```python
from sklearn.datasets import load_iris
brc=load_iris()
```

```python
brc.feature_names
```

```python
data=brc.data
target=brc.target
```

```python
plt.scatter(data[:,1],data[:,2],c=target,cmap='brg')
```

```python
from sklearn.cluster import KMeans
kms=KMeans(n_clusters=3)
#kms.fit(data[:,1:3:1])
y_pred=kms.fit_predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_pred,cmap='brg')
```

# Mean Shift

```python
from sklearn.cluster import MeanShift
msh=MeanShift(bandwidth=0.7)
y_pred=msh.fit_predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_pred,cmap='brg')
```

# Affinity Propagation

```python
from sklearn.cluster import AffinityPropagation
afp=AffinityPropagation(damping=0.9)
y_pred=afp.fit_predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_pred,cmap='brg')
```

```python
np.unique(afp.labels_).size
```

# Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
agl=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='average')
y_pred=agl.fit_predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_pred,cmap='brg')
```
