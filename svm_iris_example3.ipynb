```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
<p>We limit the study to two classes of the iris dataset</p>
```

```python
iris=load_iris()
iris.keys()
```

```python
x=iris.data[:,0:4:3]
y=iris.target

from sklearn.preprocessing import scale
x=scale(x)
```

```python
from sklearn.svm import SVC
svc=SVC(C=1.0, kernel='rbf',decision_function_shape='ovo')
plt.figure()
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
svc.fit(x,y)
```

```python
xx,yy=np.meshgrid(np.linspace(-3.0,3.0,201),np.linspace(-3.0,3.0,201))
pred=svc.predict(np.c_[xx.ravel(),yy.ravel()])
Z_svc=pred.reshape(np.shape(xx))
```

```python
plt.figure(figsize=(9,7))
plt.contourf(xx,yy,Z_svc,alpha=0.5)
plt.scatter(x.T[0],x.T[1],c=y)
plt.xlim(-3.0,3.0)
plt.ylim(-3.0,2.0)
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.title('SVC',fontsize=20)
```

```python
svc.score(x,y)
```
