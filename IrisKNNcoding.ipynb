```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=load_iris()
```

```python
iris.keys()
```

```python
iris['feature_names']
```

```python
data = iris['data']
```

```python
np.shape(data)
```

```python
plt.figure(figsize=(12,7))
plt.plot(data)
plt.title('Plot of Iris Data')
plt.xlabel('Size in cm')
plt.ylabel('Features')
plt.grid()
```

```python
plt.figure(figsize=(12,7))
plt.hist(data)
plt.title('Plot of Iris Data')
plt.xlabel('Index')
plt.ylabel('iris size')
```

```python
plt.figure(figsize=(12,7))
plt.scatter(data[:,0],data[:,3],c=iris['target'])
```

```python
plt.figure(figsize=(12,7))
plt.scatter(data[:,2],data[:,3],c=iris['target'])
```
