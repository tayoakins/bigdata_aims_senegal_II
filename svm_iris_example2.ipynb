```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
<h3>We limit the study to two classes of the iris dataset<h3>
```

```python
iris=load_iris()
iris.keys()
```

```python
x=iris.data[(iris.target==1)|(iris.target==2),0:4:2]
y=iris.target[(iris.target==1)|(iris.target==2)]

from sklearn.preprocessing import scale
x=scale(x)
```

```python
from sklearn.svm import LinearSVC
lsvc=LinearSVC(C=0.01)
plt.figure()
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
lsvc.fit(x,y)
```

```python
xx,yy=np.meshgrid(np.linspace(-3.0,3.0,201),np.linspace(-3.0,3.0,201))
pred=lsvc.predict(np.c_[xx.ravel(),yy.ravel()])
Z_lsvc=pred.reshape(np.shape(xx))
```

```python
plt.figure(figsize=(9,7))
plt.contourf(xx,yy,Z_lsvc,alpha=0.5)
plt.scatter(x.T[0],x.T[1],c=y)
plt.xlim(-3.0,3.0)
plt.ylim(-3.0,2.0)
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.title('SVC',fontsize=20)
```

```python
lsvc.score(x,y)
```
